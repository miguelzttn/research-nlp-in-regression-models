```python

df_imoveis_web_features = pd.read_csv(RAW_FOLDER / 'ImoveisWeb-Feature.csv', sep=';', encoding='latin-1', low_memory=False)
df_imoveis_web_features.head()
```

```python

df_imoveis_web_realty = pd.read_csv(RAW_FOLDER / 'ImoveisWeb-Realty.csv', sep=';', encoding='latin-1', low_memory=False)
df_imoveis_web_realty.head()
```

```python
from unidecode import unidecode


df_features = (

    pl.DataFrame(df_imoveis_web_features)

    # Handle the text
    .with_columns(

            pl.col("realty").cast(pl.Int64).alias("id_realty"),

            (
                "fg_has_"
                + pl.col('name')
                    .str.to_lowercase()
                    .str.replace_all(r"\s*\([^()]*\)", "")
                    .str.strip_chars()
                    .map_elements(unidecode)
                    .str.replace_all(r'\s+', '_') 
                + "_" 
                + pl.col('word')
                    .str.to_lowercase()
                    .str.replace_all(r"\s*\([^()]*\)", "")
                    .str.strip_chars()
                    .map_elements(unidecode)
                    .str.replace_all(r'\s+', '_')
            ).alias('combined'),

            pl.lit(1).alias('value')
    )


    # Pivot house features to columns
    .pivot(
        on='combined', 
        index='id_realty', 
        values='value', 
        aggregate_function='max',
    ).fill_null(0)

)

df_features.head()
```

```python

df_realty = (
    
    # Original DF
    pl.DataFrame(df_imoveis_web_realty)

    # Handle dtypes
    .with_columns(
        pl.col('id').cast(pl.Int64).alias('id_realty'), 
        pl.col('price').str.replace(",", ".").cast(pl.Float64).fill_null(0).alias('vl_price'),
        pl.col('pictures').cast(pl.Int64).fill_null(0).alias('n_pictures'),
        pl.col('iptu').str.replace(",", ".").cast(pl.Float64).fill_null(0).alias('vl_iptu'),
        pl.col('condominium').str.replace(",", ".").cast(pl.Float64).fill_null(0).alias('vl_condominium'),
        pl.col('totalArea').str.replace(",", ".").cast(pl.Float64).fill_null(0).alias('vl_total_area'),
        pl.col('usefulArea').str.replace(",", ".").cast(pl.Float64).fill_null(0).alias('vl_useable_area'),
        pl.col('bedroom').cast(pl.Int64).fill_null(0).alias('n_bedrooms'),
        pl.col('bathroom').cast(pl.Int64).fill_null(0).alias('n_bathrooms'),
        pl.col('garage').cast(pl.Int64).fill_null(0).alias('n_garages'),
        pl.col('years').cast(pl.Int64).fill_null(0).alias('n_years'),
        pl.col('type').cast(pl.String).alias('nm_type'),
        pl.col('title').cast(pl.Utf8).alias('nm_title'),
    )

    # Join with features
    .join(
        df_features, 
        on='id_realty', 
        how='left'
    ).with_columns(
        [
            pl.col(col).fill_null(0) 
            for col in df_features.columns 
            if col != 'id_realty'
        ]
    )

    # Add string features
    .with_columns(
        pl.col('title').cast(pl.Utf8).alias('nm_title'),
        pl.col('description').cast(pl.Utf8).alias('nm_description'),
        pl.col('address').cast(pl.Utf8).alias('nm_address'),
    )

    # Keep only columns of interest (startedd with id_, vl_, n_, nm_, fg_)
    .drop([_ for _ in df_imoveis_web_realty.columns], strict=False)

)

df_realty
```

```python

df_realty.to_pandas().to_parquet(
    ABT_FOLDER / '1_abt_realty.parquet', 
    index=False,
)
```
